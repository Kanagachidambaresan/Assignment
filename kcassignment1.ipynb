```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Demand_txn_updated.csv")
print(df.head())
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
print(df)
```

```python
df_subset = df[df['Transaction_Date'] > '2016-08-01']
print(df_subset)
```

```python
print(df_subset)
```

```python
unique_values = df['Mapped_Sales_Type'].unique()
print(unique_values)
```

```python
df_subset1 = df[df['Avg_Discount_Percent_On_Discounted_Items'] < 1.0]
sanity_check = df_subset1['Avg_Discount_Percent_On_Discounted_Items'] >= 1.0
if sanity_check.any():
    print("Warning: The new DataFrame contains values greater than or equal to 1.0 in the Avg_Discount_Percent_On_Discounted_Items column.")
else:
    print("The new DataFrame contains only values less than 1.0 in the Avg_Discount_Percent_On_Discounted_Items column.")
```

```python
df_grouped = df.groupby(['City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date']).agg({'Quantity_Sold': 'sum', 'Median_Price': 'median', 'Effective_Price': 'median'})
print(df_grouped)
```

```python
df_effective_price_not_null = df[df['Effective_Price'].notnull()]
print(df_effective_price_not_null.tail(5))
```

```python
df_chennai_delivery = df.loc[(df['City'] == 'Chennai') & (df['Mapped_Sales_Type'] == 'Delivery')]
print(df_chennai_delivery.head(5))
```

```python
df_mon_tue_wed = df[df['Day'].isin(['Mon', 'Tue', 'Wed'])]
print(df_mon_tue_wed)
```

```python
df_non_zero_discount = df.loc[df['Percent_Quantity_With_Discount'] != 0.0]
print(df_non_zero_discount)
```

```python
range_func = lambda x: 1 if x > 500 else 0.5
df['Range'] = df['Effective_Price'].apply(range_func)
print(df.head())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df['Percent_Quantity_With_Discount'] = imputer.fit_transform(df[['Percent_Quantity_With_Discount']])
print(df.head())
```
